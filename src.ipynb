```python
from scipy.integrate import quad, dblquad, tplquad
from numpy import log, sin, inf, exp
from sympy import Symbol, integrate, sin
```


## Single-variable definite integral:  ∫(x^2) dx from 0 to 1

image.png

```python
def integrand_single(x):
    return x**2

lower_limit = 0
upper_limit = 1

result, error = quad(integrand_single, lower_limit, upper_limit)

print(f"<< Single-variable definite integral result 👇 >>")
print()
print(f"(Result ± Error) = ({result} ± {error})")
```

## Another single-variable definite integral: ∫((x+1)/x^2) dx from 1 to 2

image.png

```python
def integrand_another(x):
    return (x + 1) / x**2

lower_limit = 1
upper_limit = 2

result, error = quad(integrand_another, lower_limit, upper_limit)

print(f"<< Single-variable definite integral result 👇 >>")
print()
print(f"(Result ± Error) = ({result} ± {error})")
```

## Using common functions: ∫log(sin(x)) dx from 0 to 2

image.png

```python
from numpy import log, sin

def integrand_log_sin(x):
    return log(sin(x))

lower_limit = 0
upper_limit = 2

result, error = quad(integrand_log_sin, lower_limit, upper_limit)

print(f"<< Definite integral of log(sin(x)) result 👇 >>")
print()
print(f"(Result ± Error) = ({result} ± {error})")
```

## Integral with infinity: ∫exp(-x) dx from 0 to ∞

image.png

```python
def integrand_exp(x):
    return exp(-x)

lower_limit = 0
upper_limit = inf

result, error = quad(integrand_exp, lower_limit, upper_limit)

print(f"<< Definite integral of exp(-x) from 0 to ∞ result 👇 >>")
print()
print(f"(Result ± Error) = ({result} ± {error})")
```

## Single Indefinite Integrals

### Indefinite integral of x^2

```python
x = Symbol('x')

indefinite_result_1 = integrate(x**2, x)

print(f"<< Indefinite integral of x^2 👇 >>")
print()
print(f"Result = {indefinite_result_1}")
```

## Double Integral: ∫∫ (x * y^2) dx dy over [x: 0 to 1, y: 2 to 4]

image.png

```python
def integrand_double(y, x):
    return x * y**2

inner_lower_limit = 0
inner_upper_limit = 1

outer_lower_limit = 2
outer_upper_limit = 4

result, error = dblquad(integrand_double, inner_lower_limit, inner_upper_limit, lambda x: outer_lower_limit, lambda x: outer_upper_limit)

print(f"<< Double integral result 👇 >>")
print()
print(f"(Result ± Error) = ({result} ± {error})")
```

## Triple Integral: ∫∫∫ z * (x + y + z) dx dy dz over [x: 0 to 1, y: 4 to 5, z: 0 to 1]

image.png

```python
def integrand_triple(z, y, x):
    return z * (x + y + z)

inner_lower_limit = 0
inner_upper_limit = 1

middle_lower_limit = 4
middle_upper_limit = 5

outer_lower_limit = 0
outer_upper_limit = 1

result, error = tplquad(integrand_triple, outer_lower_limit, outer_upper_limit, lambda x: middle_lower_limit, lambda x: middle_upper_limit, lambda x, y: inner_lower_limit, lambda x, y: inner_upper_limit)

print(f"<< Tripple integral result 👇 >>")
print()
print(f"(Result ± Error) = ({result} ± {error})")
```
